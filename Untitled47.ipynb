```python
import torch
import torchvision
from torch.utils.data import DataLoader
import os

# 1. Aniq manzillarni belgilaymiz (chiqqan natijangiz asosida)
train_path = "/kaggle/input/skin-cancer/datatree/train"
test_path = "/kaggle/input/skin-cancer/datatree/validation" # Sizda test o'rniga validation ekan

# 2. Transformatsiyalar
transform = torchvision.transforms.Compose([
    torchvision.transforms.Resize((256, 256)),
    torchvision.transforms.ToTensor(),
    torchvision.transforms.Normalize(
        (0.4914, 0.4822, 0.4465),
        (0.2470, 0.2435, 0.2616)
    )
])

# 3. Datasets yuklash
# ImageFolder avtomatik ravishda 'mel', 'vasc', 'nv' kabi papkalarni klass sifatida oladi
tr = torchvision.datasets.ImageFolder(train_path, transform=transform)
ts = torchvision.datasets.ImageFolder(test_path, transform=transform)

# 4. DataLoaders
tr_load = DataLoader(tr, batch_size=32, shuffle=True)
ts_load = DataLoader(ts, batch_size=32, shuffle=False)

# Natijani tekshirish
print(f"Muvaffaqiyatli!")
print(f"Klasslar soni: {len(tr.classes)} ta")
print(f"Klass nomlari: {tr.classes}")
print(f"O'qitish uchun rasmlar: {len(tr)} ta")
print(f"Test uchun rasmlar: {len(ts)} ta")
```

```python
import torch.nn as nn
import torch.nn.functional as F

class SkinCancerCNN(nn.Module):
    def __init__(self, num_classes):
        super(SkinCancerCNN, self).__init__()

        # 1-blok: Convolution + Pooling
        # Input: 3 x 256 x 256
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)
        self.pool = nn.MaxPool2d(2, 2) # Hajm 128x128 ga tushadi

        # 2-blok
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        # Poolingdan keyin hajm 64x64 ga tushadi

        # 3-blok
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        # Poolingdan keyin hajm 32x32 ga tushadi

        # To'liq bog'langan qatlamlar (Fully Connected)
        # 128 kanal * 32 * 32 o'lcham = 131072
        self.fc1 = nn.Linear(128 * 32 * 32, 512)
        self.fc2 = nn.Linear(512, num_classes)
        self.dropout = nn.Dropout(0.5)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))

        # Flat qilish (tekislash)
        x = x.view(-1, 128 * 32 * 32)

        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x

# Modelni e'lon qilish (Klasslar soni tr.classes orqali olinadi)
num_classes = len(tr.classes)
model = SkinCancerCNN(num_classes)

# Agar GPU bo'lsa, modelni GPUga o'tkazamiz
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
# 1. Modelni GPU-ga o'tkazish
model.to(device)

# 2. Training loop ichida (sizda bor ekan, lekin yana bir bor ko'ring)
for images, labels in tr_load:
    images, labels = images.to(device), labels.to(device)
    # ... qolgan qism
```

```python
epochs = 5

for epoch in range(epochs):
    model.train()
    running_loss = 0.0

    for images, labels in tr_load:
        images, labels = images.to(device), labels.to(device)

        # Gradientlarni nollash
        optimizer.zero_grad()

        # Forward pass
        outputs = model(images)
        loss = criterion(outputs, labels)

        # Backward pass va optimizatsiya
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch {epoch+1}/{epochs}, Loss: {running_loss/len(tr_load):.4f}")

print("O'qitish yakunlandi!")
```

```python
model.eval() # Modelni tekshirish rejimiga o'tkazamiz
correct = 0
total = 0

with torch.no_grad(): # Gradientlarni hisoblash shart emas (tezroq ishlaydi)
    for images, labels in ts_load:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f"Test to'plamidagi aniqlik (Accuracy): {100 * correct / total:.2d}%")
```
